## 2. Setup card

```python
def bull_head(card_num):
    """
        This function iniatial a bull head of cards in 6-nimmt! .
        return bull score for each cards
    """
    if (card_num == 55):
        return 7
    elif (card_num % 11) == 0:
        return 5
    elif (card_num % 10) == 0:
        return 3
    elif (card_num % 5) == 0:
        return 2
    else:
        return 1
```

```python
def player_point(row):
    """
        Calculate sum of bull score for each player 
    """
    total = 0
    for card in row:
        total = total + bull_head(card)
    return total
```

### 3. Logic

```python
def find_best_row(card, current_row):
    """
        Where the card could take place
    """
    best_row_index = -1 #Assume card can't place
    min_diff = 105 #Make it larger than game rule

    for i, row in enumerate(current_row):
        # print(i ,row)
        last_card = row[-1]
        if card > last_card:
            diff = card - last_card
            if diff < min_diff:
                min_diff  = diff
                best_row_index = i

    return best_row_index
```
